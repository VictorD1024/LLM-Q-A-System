```python
from datasets import Dataset
import pandas as pd
import glob

# Define the path where the CSV file is located (make sure there is a / at the end of the path)
data_path = "/content/drive/MyDrive/9900_LLM Q&A System/9900_DataSets/"

# obtain all the path of csv
file_paths = glob.glob(data_path + "*.csv")

# initialize empty DataFrame for Merge q&a pairs
all_qa_pairs = pd.DataFrame()

# load every csv and merge q&a pairs
for file in file_paths:
    try:
        chunk = pd.read_csv(file)
        print(f"Loaded {file} with {len(chunk)} rows.")
        all_qa_pairs = pd.concat([all_qa_pairs, chunk], ignore_index=True)
    except Exception as e:
        print(f"Error loading {file}: {e}")

print(f"\nTotal QA Pairs: {len(all_qa_pairs)}")
print(all_qa_pairs.head())

# Pandas DataFrame transfer to Hugging Face Dataset
dataset = Dataset.from_pandas(all_qa_pairs)
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# load pre-trained GPT-2 model and token
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# fill tokens
tokenizer.pad_token = tokenizer.eos_token
```

```python
from peft import LoraConfig, get_peft_model

# Define the configuration for LoRA
lora_config = LoraConfig(
    r=8,                          # set the rank of Low-rank matrix
    lora_alpha=32,                 # LoRA alpha paramater
    target_modules=["c_attn", "c_proj"],  # Only fine-tune specific layers
    lora_dropout=0.1,              # Dropout rate
    bias="none",
)

# Apply the LoRA configuration to the model
model = get_peft_model(model, lora_config)

#  Check the number of trainable parameters in the LoRA model
model.print_trainable_parameters()
```

```python
from transformers import GPT2Tokenizer
from datasets import Dataset, concatenate_datasets
import pandas as pd
import random

# 加载 GPT-2 分词器
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token  # 设置填充 Token 为 eos

# 随机采样数据集的函数
def sample_dataset(dataset, sample_size):
    total_size = len(dataset)
    indices = random.sample(range(total_size), min(sample_size, total_size))  # 确保采样数不超过总数据量
    return dataset.select(indices)

# 加载原始数据集并采样
sampled_dataset = sample_dataset(dataset, 100000)

# 加载 CSV 文件为 Hugging Face Dataset 格式
iot_interview_qa = pd.read_csv("/content/drive/MyDrive/9900_LLM Q&A System/iot_interview_qa.csv")
bms_interview_qa = pd.read_csv("/content/drive/MyDrive/9900_LLM Q&A System/bms_interview_qa.csv")

# 转换为 Dataset 格式
iot_interview_qa_dataset = Dataset.from_pandas(iot_interview_qa)
bms_interview_qa_dataset = Dataset.from_pandas(bms_interview_qa)

# 合并所有数据集
combined_dataset = concatenate_datasets([sampled_dataset, iot_interview_qa_dataset, bms_interview_qa_dataset])

# 定义分词函数
def preprocess_function(examples):
    inputs = [f"{q.strip()} {a.strip()}" for q, a in zip(examples['Question'], examples['Answer'])]
    model_inputs = tokenizer(
        inputs,
        max_length=512,
        truncation=True,
        padding="max_length"  # 使用最长长度的填充
    )

    model_inputs["labels"] = model_inputs["input_ids"].copy()
    return model_inputs

# 对合并的数据集进行分词，并启用缓存和多进程
tokenized_dataset = combined_dataset.map(
    preprocess_function,
    batched=True,
    num_proc=4,  # 使用 4 个进程进行并行处理
    load_from_cache_file=True  # 启用缓存
)

# 打印分词后的数据集信息
print(f"Sampled and Tokenized Dataset Size: {len(tokenized_dataset)}")
print(tokenized_dataset[0])
```

```python
# train_test_split() train and test dataset
split_dataset = tokenized_dataset.train_test_split(test_size=0.2, seed=42)


train_dataset = split_dataset['train']
eval_dataset = split_dataset['test']

# print the size of each dataset
print(f"Train Dataset Size: {len(train_dataset)}")
print(f"Eval Dataset Size: {len(eval_dataset)}")
```

```python
from transformers import GPT2LMHeadModel, Trainer, TrainingArguments

# load pre-trained GPT-2 Model
model = GPT2LMHeadModel.from_pretrained("gpt2")

# LoRA fine-tune parameter
training_args = TrainingArguments(
    output_dir="./gpt2-lora-finetuned",
    overwrite_output_dir=True,
    num_train_epochs=2,                  # epoch = 2
    per_device_train_batch_size=16,      # Batch size increases (LoRA reduces memory usage)
    per_device_eval_batch_size=8,        # Increase the validation batch size
    gradient_accumulation_steps=4,       # Increase the number of cumulative steps and simulate larger batches
    evaluation_strategy="steps",         # evaluate by steps
    eval_steps=2000,                     # every 2000 steps evalute 1 times
    save_strategy="steps",               # save every 2000 steps
    save_steps=4000,                     # Save the number of interval steps
    save_total_limit=2,
    learning_rate=1e-3,                  # learning rate = 1e-3
    warmup_steps=100,                    # reduced warmuo_steps
    weight_decay=0.01,                   # Weight decay prevents overfitting
    logging_dir="./logs",                # log
    logging_steps=200,
    fp16=True,                           # Mixed-precision training
    remove_unused_columns=False
)


# initailize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# begain training
trainer.train()
```
